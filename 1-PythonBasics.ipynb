# Complex datatypes

## Collections

### Tuples

A tuple is an ordered sequence of values (data).

Whenever I want to refer to one single data from that collection (tuple) we will use its position in the collection... First data, Second data...

```
ages = (10, 15, 20, 25)
tuple2 = (10, True, 'Ivan')
```
Index 0 -> First element in tuple

Index 1 -> Second element in tuple

Items in a tuple are not updatable:

- I cannot add new items
- I cannot remove items
- I cannot update items

### Lists

A list is an ordered sequence of values (data).

Whenever I want to refer to one single data from that collection (list) we will use its position in the collection... First data, Second data...

```
ages = [10, 15, 20, 25]
list2 = [10, True, 'Ivan']
```
Index 0 -> First element in list
Index 1 -> Second element in list

Items in a list are updatable:

- I can add new items
- I can remove items
- I can update items

### Dictionaries

A dictionary is an ordered sequence of values (data).

Whenever I want to refer to one single data from that collection (dictionary) we will use its key (id)... 

```
ages = { 'john': 10, 'mary': 15, 'paul': 20, 'lucia': 25 }
```
key 'john'  -> Value 10
key 'lucia' -> Value 25


Items in a dictionary are updatable.

## Operators

### Operators to be used with numbers

- Plus: +
- Minus: -
- Times: *
- Divided by: /
- Integer division: //
- Module (remainder): %

```

```

### Relational operators

Are operators that return a boolean (a logical values): True | False

```
number1 = 10
number2 = 20

number1 == number2
number1 != number2
number1 > number2
number1 < number2
number1 >= number2
number1 <= number2
```


### Logical operators

Are operators that return a boolean (a logical values): True | False

But in addition, they can only be applied on booleans

```
rain = True
snow = False

rain and snow
rain or snow
not rain

```

### Text operators

They return a text (string)

```

lastname = 'Osuna'
surname + lastname
surname * 3
```

### Assigment operators

